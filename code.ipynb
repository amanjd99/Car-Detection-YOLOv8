```python
import zipfile

with zipfile.ZipFile(
"cars_dataset.zip",
"r"
) as z:

    z.extractall(
    "/content/cars_dataset"
)

print("Dataset Extracted")
```

```python
yaml_content="""
path: /content/cars_dataset

train: train/images
val: valid/images
test: test/images

nc: 4

names:
  0: bike
  1: bus
  2: car
  3: truck
"""

with open(
"/content/cars_dataset/data.yaml",
"w"
) as f:

    f.write(
    yaml_content
)

print("YAML Fixed")
```

```python
model=YOLO(
"yolov8n.pt"
)

model.train(

data=
"/content/cars_dataset/data.yaml",

epochs=50,

imgsz=640,

batch=8
)
```
